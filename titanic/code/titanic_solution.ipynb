### Build Model

```python
x_data = train_df.as_matrix()
y_data = y_train_df.as_matrix()
```

```python
x_data.shape, y_data.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
clf2 = DecisionTreeClassifier(random_state=1)
eclf = BaggingClassifier(clf2, oob_score=True)
```

```python

params ={
    "n_estimators" : [10,20,30,40,50,55],
    "max_samples" : [0.5,0.6,0.7,0.8,0.9,1]
    
    }
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=eclf, param_grid=params, cv=5)
grid = grid.fit(x_data, y_data)
```

```python
result = np.concatenate((test_index.values.reshape(-1,1),
                        grid.predict(test_df.values).reshape(-1,1)), axis=1)
result[:5]
```

```python
df_submission = pd.DataFrame(result, columns=["PassengerId","Survived"])
df_submission.head()
```

```python
df_submission.to_csv("../result/submission_result.csv",index=False)
```
