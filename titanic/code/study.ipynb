#### 3.2 데이터 해석
1. 살았는지 죽었는지에 대한 결과는 1은 살았고 0은 죽었다로 표현한다. 모든 다른 변수는 잠재적인 예측변수 또는 독립적인 변수이다. 중요한 것은 많은 예측 변수들이 더 나은 모델을 만드는 것이 아니라 올바른 변수들을 만드는 것이 중요하다.
2. 탑승객ID와 티켓의 변수는 결과 즉 살았다 또는 죽었다는 것에 영향을 주지않는 무작위 값인 것으로 추정한다. 따라서 분석에 이것들은 제외될 것이다.
3. Pclass 변수는 순서자료로 사회-경제적 지위의 대리자료로 1은 상류층, 2는 중산층, 3은 하류층을 나타낸다.
4. 이름변수는명목자료유형이다. 타이틀에서 성별을, 성에서 가족의 크기, 의사나 마스터같은 타이틀에서 사회적직위를 추출하는데 사용할 수 있다. 이 변수들은 이미 존재하기 때문에 우리는 그것을 마스터와 같은 타이틀일 때 차이가 있는지에 대해서 이용할 것이다.
5. 성과 배에 탄 변수들은 명목자료유형이다. 따라서 0과 1 변수를 통해서 수리적으로 계산해 낸다.
6. 나이와 비용 변수들은 연속 정량적인 자료 유형이다. 
7. SibSp는 탑승한 관련 형제/자매의 수를 나타낸다. Parch는 탑승한 고나련 부모/자녀 수를 나타낸다. 둘 다 별개의 정량적 데이터 유형이며 이것은 특징으로 사용되어 패밀리의 크기를 만들 수 있으며 단독 변수이다.
8. 객실 변수는 명목자료 유형인데 배에서 대략적인 위치와 갑판 수준으로 부터 사회적지위를 나타낸다. 그러나 많은 null값에 의해서 이것은 분석에서 제외시킨다.

#### 3.21 Data Cleaning : 수정, 완료, 생성, 및 변환
1. Correcting : 데이터를 검토한 결과, 잘못된 데이터 입력 이나 허용할 수 없는 데이터 입력은 없다. 게다가 Age와 Fare에서 잠재적 특이치들을 가지고 있을 수 있는것을 알고 있따. 그러나 그것들은 합리적인 가격이기 때문에 우리는 데이터에 포함해야할지 제외해야할지 결정하기 위해 분석 후에 결정해야한다. 주목해야할 것은, 만약 그것들이 합리적인 변수가 아니라 예를 들어 나이가 800살 일 경우에는 80살로 고치는 것은 아마도 안전한 결정일 것이다. 그러나 정확한 모델을 만드는 것이 필요할 수 있기 때문에 데이터를 원래 값에서 수정할 때 주의를 기울여야 한다.
2. Completing : Age, Cabin, Embarked 변수에 null 값 또는 누락된 데이터가 있다. 일부 알고리즘은 null값에 의해서 처리하는 방법이 없어 실패할 수 있기 때문에 누락된 값은 좋지 않을 수 있다. 따라서 여러 모델을 비교하고 대조할 것이기 때문에 모델링을 시작하기 전에 수정하는 것이 중요하다. 레코드를 삭제하거나 또는 누락된 값을 합리적인 값으로 채워넣는 흔한 방법이 두가지가 있다. 정말 분완전한 값으로 나타내지 않는 한, 많은 비율의 레코드를 삭제하는 것은 권장하지 않는다. 그 대신, 누락된 값을 다른 값으로 넣어주는 것이 최선이다. 정성적 데이터에 대한 기본 방법은 최빈값를 사용하는 것이다. 정량데이터에 대한 기본 방법론은 평균, 중위 또는 평균+무작위 표준편차를 사용하여 귀속한다. 중간 방법론은 계급별 평균 연령 또는 운임과 사회직위에 의한 탑승 항만과 같은 특정 기준에 기초한 기본 방법론을 사용하는 것이다. 더 복잡한 방법론 들이 있지만, 배치하기 전에 기본 모델과 비교하여 복잡성이 진정으로 가치를 추가하는지 여부를 결정해야 한다. 이 데이터의 경우, 연령은 중위값으로, cabin은 삭제하며, embark는 최빈값으로 해준다. 이후 모델의 반복은 모델의 정확도를 향상시키는지를 결정하기 위해 이 결정을 수정할 수 있다.
3. Creating : feature engineering은 결과를 예측하기 위해서 새로운 신호가 제공된다면 새로운 특징을 만들기 위해서 기존의 특징을 사용하는 것이다. 이 데이터셋은 제목 기능을 생성하여 생존에 역할을 했는지 여부를 판단할 것이다.
4. Converting : 마지막으로 정말 중요한 것은 Formatting이다. 날짜나 통화 형식으로 변환되어 있지 않으면 계산하기에 어렵다. 따라서 데이터의 유형을 범주형 더미로 변환할 것이다.

#### 3.22 Clean Data

```python
# Completing
for dataset in data_cleaner:    
    #complete missing age with median
    dataset['Age'].fillna(dataset['Age'].median(), inplace = True) # Age가 없는 값은 중위수로 채워줌

    #complete embarked with mode
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace = True)  # Embarked가 없는 값은 최빈값으로 채워줌

    #complete missing fare with median
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace = True) # Fare가 없는 값은 중위수로 채워줌

#delete the cabin feature/column and others previously stated to exclude in train dataset
drop_column = ['PassengerId','Cabin', 'Ticket']
data1.drop(drop_column, axis=1, inplace = True)

print(data1.isnull().sum())
print("-"*100)
print(data_val.isnull().sum())
```

```python
# Create
for dataset in data_cleaner:    
    # 형재자매수+배우자+부모+자녀+나자신 = 가족 수
    dataset['FamilySize'] = dataset ['SibSp'] + dataset['Parch'] + 1 

    #혼자인 사람은 1 가족이 있는 사람은 0으로 바꿔줌
    dataset['IsAlone'] = 1 
    dataset['IsAlone'].loc[dataset['FamilySize'] > 1] = 0 

    # 'Mr', 'Mrs', 'Miss', 'Master', 'Ms', 'Col', 'Rev', 'Dr', 'Dona'로 titile이란 변수 생성
    dataset['Title'] = dataset['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]

    # qcut을 이용해서 Fare의 값을 4개 범위로 나눠 범위를 Farebin으로 만들어줌
    dataset['FareBin'] = pd.qcut(dataset['Fare'], 4)

    # cut을 이용해서 Age의 값을 5개의 범위로 나눠 범위를 Agebin으로 만들어줌
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(int), 5)

# qcut과 cut의 차이 : qcut은 빈이 값의 빈도에 따라 범위를 나누지만 cut은 값 자체에 따라 균등하게 간격을 두고 나눔

#cleanup rare title names
#print(data1['Title'].value_counts())
stat_min = 10 
# Title 이 10개도 안되는 부분을 True/False로 series형태로 나타냄
title_names = (data1['Title'].value_counts() < stat_min)

# apply 함수를 이용해 titile name이 True면 그값을 넣고 아니면 Misc를 넣어준다
data1['Title'] = data1['Title'].apply(lambda x: 'Misc' if title_names.loc[x] == True else x)
print(data1['Title'].value_counts())
print("-"*100)


#preview data again
data1.info()
data_val.info()
data1.sample(10)
```

#### 3.23 Convert Formats
: 데이터를 수학분석을 하기 위해서 one-hot encoding으로 변환 해주도록 한다.

```python
#Convert

label = LabelEncoder()
for dataset in data_cleaner:
    # Sex, Embarked, Title, AgeBin, FareBin을 Label로 바꿔준다
    dataset['Sex_Code'] = label.fit_transform(dataset['Sex'])
    dataset['Embarked_Code'] = label.fit_transform(dataset['Embarked'])
    dataset['Title_Code'] = label.fit_transform(dataset['Title'])
    dataset['AgeBin_Code'] = label.fit_transform(dataset['AgeBin'])
    dataset['FareBin_Code'] = label.fit_transform(dataset['FareBin'])


# 목표인 Survived를 타겟으로 설정한다.
Target = ['Survived']

# 그래프로 확인한 변수명들과 알고리즘에 들어갈 변수명을 정리해준다.
data1_x = ['Sex','Pclass', 'Embarked', 'Title','SibSp', 'Parch', 'Age', 'Fare', 'FamilySize', 'IsAlone'] #pretty name/values for charts
data1_x_calc = ['Sex_Code','Pclass', 'Embarked_Code', 'Title_Code','SibSp', 'Parch', 'Age', 'Fare'] #coded for algorithm calculation
data1_xy =  Target + data1_x
print('Original X Y: ', data1_xy, '\n')


# 연속변수를 제거할 X변수를 정의한다.
data1_x_bin = ['Sex_Code','Pclass', 'Embarked_Code', 'Title_Code', 'FamilySize', 'AgeBin_Code', 'FareBin_Code']
data1_xy_bin = Target + data1_x_bin
print('Bin X Y: ', data1_xy_bin, '\n')


# get_dummies를 이용해 one-hot encoding으로 변환해준다
data1_dummy = pd.get_dummies(data1[data1_x])
data1_x_dummy = data1_dummy.columns.tolist()
data1_xy_dummy = Target + data1_x_dummy
print('Dummy X Y: ', data1_xy_dummy, '\n')



data1_dummy.head()
```

```python
#Machine Learning Algorithm (MLA) Selection and Initialization
MLA = [
    #Ensemble Methods
    ensemble.AdaBoostClassifier(),
    ensemble.BaggingClassifier(),
    ensemble.ExtraTreesClassifier(),
    ensemble.GradientBoostingClassifier(),
    ensemble.RandomForestClassifier(),

    #Gaussian Processes
    gaussian_process.GaussianProcessClassifier(),
    
    #GLM
    linear_model.LogisticRegressionCV(),
    linear_model.PassiveAggressiveClassifier(),
    linear_model.RidgeClassifierCV(),
    linear_model.SGDClassifier(),
    linear_model.Perceptron(),
    
    #Navies Bayes
    naive_bayes.BernoulliNB(),
    naive_bayes.GaussianNB(),
    
    #Nearest Neighbor
    neighbors.KNeighborsClassifier(),
    
    #SVM
    svm.SVC(probability=True),
    svm.NuSVC(probability=True),
    svm.LinearSVC(),
    
    #Trees    
    tree.DecisionTreeClassifier(),
    tree.ExtraTreeClassifier(),
    
    #Discriminant Analysis
    discriminant_analysis.LinearDiscriminantAnalysis(),
    discriminant_analysis.QuadraticDiscriminantAnalysis(),

    
    #xgboost: http://xgboost.readthedocs.io/en/latest/model.html
    XGBClassifier()    
    ]



#split dataset in cross-validation with this splitter class: http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.ShuffleSplit.html#sklearn.model_selection.ShuffleSplit
#note: this is an alternative to train_test_split
cv_split = model_selection.ShuffleSplit(n_splits = 10, test_size = .3, train_size = .6, random_state = 0 ) # run model 10x with 60/30 split intentionally leaving out 10%

#create table to compare MLA metrics
MLA_columns = ['MLA Name', 'MLA Parameters','MLA Train Accuracy Mean', 'MLA Test Accuracy Mean', 'MLA Test Accuracy 3*STD' ,'MLA Time']
MLA_compare = pd.DataFrame(columns = MLA_columns)

#create table to compare MLA predictions
MLA_predict = data1[Target]

#index through MLA and save performance to table
row_index = 0
for alg in MLA:

    #set name and parameters
    MLA_name = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Name'] = MLA_name
    MLA_compare.loc[row_index, 'MLA Parameters'] = str(alg.get_params())
    
    #score model with cross validation: http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html#sklearn.model_selection.cross_validate
    cv_results = model_selection.cross_validate(alg, data1[data1_x_bin], data1[Target], cv  = cv_split)

    MLA_compare.loc[row_index, 'MLA Time'] = cv_results['fit_time'].mean()
#     MLA_compare.loc[row_index, 'MLA Train Accuracy Mean'] = cv_results['train_score'].mean()
    MLA_compare.loc[row_index, 'MLA Test Accuracy Mean'] = cv_results['test_score'].mean()   
    #if this is a non-bias random sample, then +/-3 standard deviations (std) from the mean, should statistically capture 99.7% of the subsets
    MLA_compare.loc[row_index, 'MLA Test Accuracy 3*STD'] = cv_results['test_score'].std()*3   #let's know the worst that can happen!
    

    #save MLA predictions - see section 6 for usage
    alg.fit(data1[data1_x_bin], data1[Target])
    MLA_predict[MLA_name] = alg.predict(data1[data1_x_bin])
    
    row_index+=1

    
#print and sort table: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sort_values.html
MLA_compare.sort_values(by = ['MLA Test Accuracy Mean'], ascending = False, inplace = True)
MLA_compare
#MLA_predict
```
